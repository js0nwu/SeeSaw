```python
folderPath = "/Users/gareyes/Downloads/TLX/"

categories = ['Mental Demand', 'Physical Demand', 'Temporal Demand', 'Performance', 'Effort', 'Frustration']

for root, dirs, files in os.walk(folderPath, topdown=False):
    for name in files:
        fileName = os.path.join(root, name)
        if fileName.endswith(".csv"):
            with open(fileName) as f:
                overallRating = []
                lis = f.readlines()
                for line in lis:
                    if line.strip():
                        lineSplit = line.split(",")
                        if any(category in line for category in categories):                        
                            categoryName = lineSplit[0]
                            rawRating = lineSplit[1]
                            tally = lineSplit[2]
                            print categoryName + "," + str(rawRating) + "," + str(float(tally)/15)
                            overallRating.append((float(tally)/15) * float(rawRating))
                        if "Subject name:" in line:
                            participant = lineSplit[1][:-1]
                        if "Task text:" in line:
                            task = lineSplit[1][1:-2]
                print fileName[:-1] + "," + participant + "," + task + "," + str(np.sum(overallRating))
```
